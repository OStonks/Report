Объединим две выписки в одну таблицу

```python
result1 = pd.concat([v122, v123], ignore_index=True)
result1
```

Теперь добавим столбец 'Сумма операции', однако в нем встречаются посторонние знаки, которые не дают возможность перевести строковые значения в float, а потому мы избавимся от них

```python
# Удаляем все, кроме цифр и запятых
result1['Сумма операции'] = result1['Сумма операции'].str.replace(r'[^\d,]', '', regex=True)  
# Заменяем запятые на точки и преобразуем в float
result1['Сумма операции'] = result1['Сумма операции'].str.replace(',', '.', regex=True).astype('float')  
#создаем столбец Комиссия
result1['Комиссия'] = round(result1['Сумма операции']*0.0125, 2)
#выводим результат
result1
```

```python
#создаем столбец Сумма к переводу
result1['Сумма к переводу'] = result1['Сумма операции'] - result1['Комиссия']
#выводим результат
result1
```

```python
#объединяем таблицы и добавляем столбец Код авторизации
uni['Код авторизации']= uni['AuthCode']
data= result1.merge(uni[['Код авторизации','OrderID']], on='Код авторизации')
```
